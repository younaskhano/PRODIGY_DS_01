```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("population-data.csv")
```

```python
data.head(266)
```

```python
data.describe
```

```python
#List of countries for the graph
countries_to_plot = ['United States','Canada','Germany','Pakistan','India','Japan']

#Filtering the data frame
filtered_data = data[data['Country Name'].isin(countries_to_plot)]

#Creating the bar chart
plt.bar(filtered_data['Country Name'], filtered_data['2022'])
plt.xlabel ('Country Name',fontsize=12)
plt.ylabel ('Population in 2022',fontsize=12)
plt.title ('Population of selected countries in 2022',fontsize=18)
plt.xticks (rotation=45, ha='right',fontsize=10) #Rotating the x-axis labels for readability
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
